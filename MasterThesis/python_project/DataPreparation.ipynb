```python
import numpy as np
import pandas as pd
import math
import yfinance as yf
```

```python
nse_df = pd.read_csv("data/NSE-StockHistory.csv")
```

```python
nse_df.head()
```

```python
ticker_set = set()  # Create an empty ticker set that will hold all the tickers from 2000 to 2024
for year in range(2000,2024):
    for ticker in nse_df[str(year)].values:
        if isinstance(ticker, str):
            ticker_set.add(ticker)        
```

```python
print(ticker_set)
```

```python
sorted(ticker_set)
```

```python
nifty_50_symbols = sorted(ticker_set)

# Define the date range
start_date = "2000-01-01"
end_date = "2024-12-31"

# Create an empty DataFrame to store returns
returns_df = pd.DataFrame()
vol_df = pd.DataFrame()
raw_df = pd.DataFrame()

# Fetch data and calculate returns for each symbol
for symbol in nifty_50_symbols:
    # Download historical data
    data = yf.download(symbol, start=start_date, end=end_date)
    # Add to raw both close and volume
    raw_df[symbol]=data['Close']
    # Calculate daily returns
    returns = data['Close'].pct_change()
    # remove possible nonsensical returns
    returns[returns<-1] = -1
    vol = data['Volume']
    # Add returns to the DataFrame
    returns_df[symbol] = returns
    vol_df[symbol] = vol

# Set the index name
returns_df.index.name = "Date"
vol_df.index.name = "Date"
raw_df.index.name = "Date"
```
